```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock = pd.read_csv("APLE_stock.csv")
stock.head(20)
```

```python
stock.to_csv("APLE_stock.csv", index=False)
stock.head(50)
```

```python
Q1 = stock['Volume'].quantile(0.25)
Q3 = stock['Volume'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
volume_outliers = stock[(stock['Volume'] < lower) | (stock['Volume'] > upper)]
print(volume_outliers)
```

```python
#Describe the dataset with basic statistics: mean, median, max, min, and standard deviation for columns like 'Open', 'Close', and 'Volume'.


stats = stock[['Open','Close','Volume']].agg(['mean','median','std','min','max'])
display(stats)
```

```python
stock.plot()
```

```python
# 1. Calculate 20-day moving average
stock['MA20'] = stock['Close'].rolling(window=20).mean()

# 2. Detect outliers (Close > mean + 1.5*std)
mean_close = stock['Close'].mean()
std_close = stock['Close'].std()
outliers = stock[stock['Close'] > mean_close + 1.5*std_close]

# 3. visuals
plt.figure(figsize=(12,6))
plt.plot(stock['Close'], label='Close Price', color='blue')
plt.plot(stock['MA20'], label='20-Day MA', color='red', linestyle='--')
plt.scatter(outliers.index, outliers['Close'], color='orange', label='Outliers', s=50)
plt.title('APLE Stock Closing Price (2024)')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Investigate the trend of the stock's closing price over time by plotting a line graph.


# Plot APLE closing price trend
plt.figure(figsize=(12,6))
plt.plot(stock['Close'], color='blue', label='Close Price')
plt.title('APLE Stock Closing Price (2024)')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Investigate correlations between numerical columns, especially between 'Open' and 'Close' prices, and visualize the correlation matrix.

# 1. Calculate correlation matrix
corr_matrix = stock[['Open','Close']].corr()

# 2. Display correlation matrix
print(corr_matrix)

# 3. Visualize as a heatmap
plt.figure(figsize=(5,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of APLE Stock Data')
plt.show()
```

```python
#Is there a relationship between the volume of trades and stock price movements? Create a scatter plot or line chart comparing 'Volume' vs. 'Close' price.

# 1. Load the dataset
df = pd.read_csv('APLE_stock.csv')

# 2. Preview the dataset
print(df.head())
# 3. Scatter plot: Volume vs Close price
plt.figure(figsize=(10,6))
plt.scatter(df['Volume'], df['Close'], alpha=0.5)
plt.title('Volume vs Close Price')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()

# 4. Line chart: Close price and Volume over time
fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price', color='blue')
ax1.plot(df['Date'], df['Close'], color='blue', label='Close Price')

ax2 = ax1.twinx()  # Create a second y-axis for Volume
ax2.set_ylabel('Volume', color='red')
ax2.plot(df['Date'], df['Volume'], color='red', alpha=0.5, label='Volume')

plt.title('Stock Close Price and Volume Over Time')
plt.show()
```

```python
#Calculate the 30-day Simple Moving Average (SMA) and 100-day SMA for the closing price.

# 1. Load the dataset
df = pd.read_csv('APLE_stock.csv')

# 2. Calculate 30-day and 100-day Simple Moving Averages (SMA)
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_100'] = df['Close'].rolling(window=100).mean()

# 3. Display first few rows
print(df[['Date', 'Close', 'SMA_30', 'SMA_100']].head(15))

# 4. Plot Close price and SMAs
plt.figure(figsize=(14,7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.plot(df['Date'], df['SMA_30'], label='30-Day SMA', color='orange')
plt.plot(df['Date'], df['SMA_100'], label='100-Day SMA', color='green')

plt.title('Close Price with 30-Day and 100-Day SMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Create buy/sell signals:
 #   a. Buy when the 30-day SMA crosses above the 100-day SMA.
  #  b. Sell when the 30-day SMA crosses below the 100-day SMA.


df = pd.read_csv('APLE_stock.csv')

# 1. Calculate SMAs
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_100'] = df['Close'].rolling(window=100).mean()

# 2. Create buy/sell signals
df['Signal'] = 0  # default no action
df['Signal'] = df.apply(lambda row: 1 if row['SMA_30'] > row['SMA_100'] else -1, axis=1)
df['Position'] = df['Signal'].diff()  # detect crossovers

# 3. Buy/Sell markers
buy_signals = df[df['Position'] == 2]   # when SMA_30 crosses above SMA_100
sell_signals = df[df['Position'] == -2] # when SMA_30 crosses below SMA_100

# 4. Plot
plt.figure(figsize=(14,7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue', alpha=0.5)
plt.plot(df['Date'], df['SMA_30'], label='30-Day SMA', color='orange', linewidth=1.5)
plt.plot(df['Date'], df['SMA_100'], label='100-Day SMA', color='green', linewidth=1.5)

# Plot Buy/Sell signals
plt.scatter(buy_signals['Date'], buy_signals['Close'], label='Buy Signal', marker='^', color='green', s=100)
plt.scatter(sell_signals['Date'], sell_signals['Close'], label='Sell Signal', marker='v', color='red', s=100)

plt.title('Buy & Sell Signals (SMA Crossovers)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()

# 8. Preview signals
print(df[['Date', 'Close', 'SMA_30', 'SMA_100', 'Signal', 'Position']].tail(15))
```

```python
#Visualize the buy/sell signals on the same plot as the closing price and the two SMAs.


# 2. Load dataset
df = pd.read_csv('APLE_stock.csv')

# 3. Calculate SMAs
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_100'] = df['Close'].rolling(window=100).mean()

# 4. Create Buy/Sell signals
df['Signal'] = 0
df.loc[df['SMA_30'] > df['SMA_100'], 'Signal'] = 1
df.loc[df['SMA_30'] <= df['SMA_100'], 'Signal'] = -1
df['Position'] = df['Signal'].diff()

# 5. Extract Buy/Sell points
buy_signals = df[df['Position'] == 2]
sell_signals = df[df['Position'] == -2]

# 6. Visualization
plt.figure(figsize=(14,7))

# Plot Close price and SMAs
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue', alpha=0.6)
plt.plot(df['Date'], df['SMA_30'], label='30-Day SMA', color='orange', linewidth=1.5)
plt.plot(df['Date'], df['SMA_100'], label='100-Day SMA', color='green', linewidth=1.5)

# Plot Buy signals
plt.scatter(buy_signals['Date'], buy_signals['Close'], marker='^', color='green', label='Buy Signal', s=100)

# Plot Sell signals
plt.scatter(sell_signals['Date'], sell_signals['Close'], marker='v', color='red', label='Sell Signal', s=100)

# Add title and labels
plt.title('Stock Price with Buy/Sell Signals (SMA Crossovers)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```
